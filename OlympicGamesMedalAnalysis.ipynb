```python
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt
from ydata_profiling import ProfileReport
```

```python
path = r'C:\Users\Begra\OneDrive\Documents\GitHub\OlympicGamesMedal'
all_files = glob.glob(os.path.join(path, "/*.csv"))
```

```python
# Print the path to verify
print("Path:", path)

# Use glob to get all CSV files in the directory
all_files = glob.glob(os.path.join(path, "*.csv"))

# Print the list of files to verify
print("Files found:", all_files)

# Check if any files were found
if not all_files:
    print("No CSV files found in the directory.")
else:
    # Create an empty list to store the DataFrames
    dfs = []

    # Loop through the list of files and read each one into a DataFrame
    for file in all_files:
        df = pd.read_csv(file)
        dfs.append(df)

    # Concatenate all the DataFrames into a single DataFrame
    combined_df = pd.concat(dfs, ignore_index=True)

    # Print the combined DataFrame
    print(combined_df)
```

```python
profile = ProfileReport(combined_df)
profile
```

```python
grouped_df_gold = combined_df.groupby('NOC')['Gold'].sum().reset_index()

grouped_df_gold = grouped_df_gold.sort_values(by='Gold', ascending=False)

plt.figure(figsize=(30, 6))
plt.bar(grouped_df_gold['NOC'], grouped_df_gold['Gold'], color='skyblue')
plt.xlabel('NOC')
plt.ylabel('Gold')
plt.title('Countries with Highest Gold Medals')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
